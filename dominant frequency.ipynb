```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Arial"
plt.rcParams.update({'font.size': 24})
plt.rcParams['xtick.major.pad']='12'
plt.rcParams['ytick.major.pad']='12'
```

```python
# P velocity of upper acoustic layer [m/s].
alpha1=1420.0
# P velocity range of elastic half space [m/s].
alpha2=np.linspace(1420.1,1430.0,1000)
# Thickness of layer [m].
h=5.0
```

```python
# Compute critical angles.
phi_pc=np.arcsin(alpha1/alpha2)
# Compute peak frequencies for fundamental mode.
f0=alpha1/(4.0*h*np.cos(phi_pc))
```

```python
fig=plt.figure(figsize=(10,10))
plt.plot(alpha2/alpha1,f0,'k',linewidth=2)
plt.plot(alpha2/alpha1,3.0*f0,'--k',linewidth=1)

plt.minorticks_on()
plt.grid(which='major',color='k',linewidth=1.0)
plt.grid(which='minor',color='k',linewidth=0.5)
plt.xlabel('alpha2/alpha1 [1]',labelpad=20)
plt.ylabel('dominant frequency [Hz]',labelpad=20)
plt.xlim([1.0,1.0065])
plt.ylim([0.0,3000.0])
plt.tight_layout()
plt.savefig('f_dominant.pdf',format='pdf')
plt.show()
```
